# step 1. full-stack 업무 리스트 크롤링
- command+option+i : 맥 페이지 소스 단축키
- div class='content' > div class='jobs-container' > section class='jobs' > article > ul > li

```python
import cloudscraper # 크롤링 우회 모듈 cloudscraper
from bs4 import BeautifulSoup

scraper = cloudscraper.create_scraper()

url = 'https://weworkremotely.com/categories/remote-full-stack-programming-jobs#job-listings'

response = scraper.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

# 찾고자하는 section의 class/id 입력 -> 모든 list 출력 (마지막 제외)
# 마지막 리스트 : 'back to all jobs'
jobs = soup.find('section', id='category-2').find_all('li')[:-1]

print(jobs)
```

# step 2. 직무명, 지역, 회사명 출력

```python
for job in jobs :
    header = job.find('h4', class_='new-listing__header__title').text # 직무명 출력
    region = job.find_all('p', class_='new-listing__categories__category') # 지역 출력
    company = job.find('p', class_='new-listing__company-name').text # 회사명 출력
    
    if len(region) > 0 :
        region = job.find_all('p', class_='new-listing__categories__category')[-1].text # nontype이 있으므로 if문에서 text 추출
    else :
        region = 'no region'
        
    print(header, region, company,'-------\n')
    
```

# step3. dictionary 생성

```python
all_jobs = [] # 리스트 생성

for job in jobs :
    header = job.find('h4', class_='new-listing__header__title').text # 직무명 출력
    region = job.find_all('p', class_='new-listing__categories__category') # 지역 출력
    company = job.find('p', class_='new-listing__company-name').text # 회사명 출력
    
    if len(region) > 0 :
        region = job.find_all('p', class_='new-listing__categories__category')[-1].text # nontype이 있으므로 if문에서 text 추출
    else :
        region = 'no region'

    # 딕셔너리 생성
    job_data = {
        'header': header,
        'region': region,
        'company': company}
    all_jobs.append(job_data) # all_jobs 리스트에 추가

print(all_jobs)
```

```python
# url 추가

all_jobs = [] # 리스트 생성

for job in jobs :
    header = job.find('h4', class_='new-listing__header__title').text 
    region = job.find_all('p', class_='new-listing__categories__category') 
    company = job.find('p', class_='new-listing__company-name').text 

    if len(region) > 0 :
        region = job.find_all('p', class_='new-listing__categories__category')[-1].text 
    else :
        region = 'no region'


    try:
        url = job.find('div', class_='tooltip--flag-logo').next_sibling['href'] # 두번째 anchor을 검색해야하므로 첫번째 anchor의 다음으로 설정
    except (AttributeError, TypeError): # error는 예외처리
        url = None


    job_data = {
        'header': header,
        'region': region,
        'company': company,
        'url': f'https://weworkremotely.com/{url}'} # 하이퍼링크 형태로 변환
    all_jobs.append(job_data)

print(all_jobs)
```

# step4. pagination
- div class='pagination' -> span class='page'

```python
import cloudscraper # 크롤링 우회 모듈 cloudscraper
from bs4 import BeautifulSoup

scraper = cloudscraper.create_scraper()

all_jobs = []

def scrape_page(url):
    print(f'Scrapping {url}...')
    response = scraper.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    jobs = soup.find('section', class_='jobs').find_all('li')[:-1]

    for job in jobs :
        header = job.find('h4', class_='new-listing__header__title').text 
        region = job.find_all('p', class_='new-listing__categories__category') 
        company = job.find('p', class_='new-listing__company-name').text 
    
        if len(region) > 0 :
            region = job.find_all('p', class_='new-listing__categories__category')[-1].text 
        else :
            region = 'no region'
    
    
        try:
            url = job.find('div', class_='tooltip--flag-logo').next_sibling['href'] 
        except (AttributeError, TypeError): 
            url = None
    
    
        job_data = {
            'header': header,
            'region': region,
            'company': company,
            'url': f'https://weworkremotely.com/{url}'} 
        all_jobs.append(job_data)

# 페이지 수 가져오는 함수
def get_pages(url):
    response = scraper.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    return len(soup.find('div', class_='pagination').find_all('span', class_='page')) # page 부분 크롤링, len 반환


total_pages = get_pages('https://weworkremotely.com/remote-full-time-jobs?page=1') # 해당 url의 total page 수

for x in range(total_pages): # 페이지 수만큼 반복
    url = f'https://weworkremotely.com/remote-full-time-jobs?page={x+1}' # 숫자는 0부터 세므로  +1 해줌
    scrape_page(url)

print(len(all_jobs)) # 일자리 총 개수
```
