# step4. pagination
- div class='pagination' -> span class='page'

```python
import cloudscraper # í¬ë¡¤ë§ ìš°íšŒ ëª¨ë“ˆ cloudscraper
from bs4 import BeautifulSoup

scraper = cloudscraper.create_scraper()

all_jobs = []

def scrape_page(url):
    print(f'Scrapping {url}...')
    response = scraper.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    jobs = soup.find('section', class_='jobs').find_all('li')[:-1]

    for job in jobs :
        header = job.find('h4', class_='new-listing__header__title').text 
        region = job.find_all('p', class_='new-listing__categories__category') 
        company = job.find('p', class_='new-listing__company-name').text 
    
        if len(region) > 0 :
            region = job.find_all('p', class_='new-listing__categories__category')[-1].text 
        else :
            region = 'no region'
    
    
        try:
            url = job.find('div', class_='tooltip--flag-logo').next_sibling['href'] 
        except (AttributeError, TypeError): 
            url = None
    
    
        job_data = {
            'header': header,
            'region': region,
            'company': company,
            'url': f'https://weworkremotely.com/{url}'} 
        all_jobs.append(job_data)

# íŽ˜ì´ì§€ ìˆ˜ ê°€ì ¸ì˜¤ëŠ” í•¨ìˆ˜
def get_pages(url):
    response = scraper.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    return len(soup.find('div', class_='pagination').find_all('span', class_='page')) # page ë¶€ë¶„ í¬ë¡¤ë§, len ë°˜í™˜


total_pages = get_pages('https://weworkremotely.com/remote-full-time-jobs?page=1') # í•´ë‹¹ urlì˜ total page ìˆ˜

for x in range(total_pages): # íŽ˜ì´ì§€ ìˆ˜ë§Œí¼ ë°˜ë³µ
    url = f'https://weworkremotely.com/remote-full-time-jobs?page={x+1}' # ìˆ«ìžëŠ” 0ë¶€í„° ì„¸ë¯€ë¡œ  +1 í•´ì¤Œ
    scrape_page(url)

print(len(all_jobs)) # ì¼ìžë¦¬ ì´ ê°œìˆ˜
```
